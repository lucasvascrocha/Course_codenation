```python
# todas os features que batem com o test.csv que acredito que tenha relação
dftudo = pd.DataFrame(train[['CO_UF_RESIDENCIA', 'NU_IDADE', 'TP_SEXO', 'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'IN_TREINEIRO', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'TP_LINGUA','TP_STATUS_REDACAO', 'NU_NOTA_REDACAO', 'Q001' ,'Q002', 'Q006', 'Q024', 'Q025', 'Q026', 'Q047' ]])
dftudo
```

```python
# todos os features de notas apenas
dfnotas = pd.DataFrame(dftudo[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO', 'NU_NOTA_MT']])
dfnotas
```

```python
# trocando NaN por 0
dfnotas = dfnotas.fillna(0)
dfnotas
```

```python
# Modelo de regressão Linear
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = dfnotas.drop(['NU_NOTA_MT'], axis=1)
y = dfnotas['NU_NOTA_MT']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# mdoel training and building
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_jobs=-1, n_estimators=50, verbose=3)
rfr.fit(X_train,y_train)
```

```python
#fit model or training data
lm.fit(X_train, y_train)
```

```python
# model evaluation
# print intercepts
print(lm.intercept_)
```

```python
# prints coeficinetes de cada uma das colunas X_train
lm.coef_
```

```python
#  Inlfuência de cada nota na nota de matemática, Licenciatura
coeffecients = pd.DataFrame(lm.coef_.T,X.columns.T)
coeffecients.columns = ['Coeffecient']
coeffecients
```

```python
# predictions off test set
predictions = lm.predict(X_test)
predictions
```

```python
predictions = rfr.predict(X_test)
```

```python
y_test
```

```python
plt.scatter(y_test, predictions)
```

```python
sns.distplot((y_test - predictions))
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
